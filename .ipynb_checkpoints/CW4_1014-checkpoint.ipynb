2. replace data

    - replace it based on other functions

Whole columns should be dropped only if most entries in the column are missing. In our dataset, none of the columns are sparse enough to drop entirely. We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

## Data Standardization

Standardization is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.

Example: Transform **mpg** to **L/100km**
- In our dataset, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard 
- We will need to transform mpg into L/100km. The formula for unit conversion is: **L/100km = 235 / mpg**

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235 / df['city-mpg']
```

```python
# check the transformed data 
df.head()
```

## Data Normalization 

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include min-max and z-score approaches. 

To demonstrate normalization, let's say we want to scale the column <code>length, width, height</code> using the min-max approach. This will rescale the columns to the same range [0,1]


Nnormalize the column <code>length</code>

```python
# normalize the original value using min-max approach: (original-min)/(max-min)
df['length'] = (df['length'] - df['length'].min()) / (df['length'].max() - df['length'].min())
```

```python
# show the scaled column
df['length'].head()
```

## Binning

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

Example: 

In our dataset, "price" is a real valued variable ranging from 5118 to 45400. We can categorize the cars into **high**, **medium**, and **low** priced cars (3 types). 

We will use the Pandas method 'cut' to segment the 'price' column into 3 bins

Lets plot the histogram of price, to see what the distribution looks like.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure() 
```

```python
# Visualize the distribution of car prices
sns.histplot(df["price"])
```

We would like 3 bins of equal size bandwidth so we use numpy's <code>linspace(start_value, end_value, numbers_generated</code> function. 

- We want to include the min and max values of price, so we set start_value as the min and end_value as the max of the price column. 
- Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.

```python
# find the value dividers 
bins = np.linspace(df['price'].min(), df['price'].max(), 4)
bins
```

```python
# Set group names for the bins
group_names = ['Low', 'Meidum', 'High']
```

```python
# We apply the function "cut" the determine what each value of car price belongs to. 
df['price-binned'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)
```

```python
df['price-binned'].head()
```

Lets see the number of vehicles in each bin.


```python
# Lets see the number of vehicles in each bin
df['price-binned'].value_counts()
```

## Indicator variable (or dummy variable) 

An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 

Example: We see the column "fuel-type" has two unique values, "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.

We will use the panda's method 'get_dummies' to assign numerical values to different categories of fuel type. 

```python
# get the number of unique values in a given column
df['fuel-type'].value_counts()
```

Get indicator variables and save them to a data frame "dummy_variable_1". For rows with <code>fuel-type</code> being 'gas', the <code>gas</code> column will be '1' and <code>diesel</code> will be '0'. Similarly for rows with <code>fuel-type</code> bing 'diesel', we will have 1 for <code>diesel</code> and 0 for <code>gas</code>.


```python
# One-hot encoding
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

Rename these two columns for clarity 

```python
# Rename these two columns 
dummy_variable_1.rename(columns = {'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True )
```

```python
dummy_variable_1.head()
```

Now, merge the dataframe <code>dummy_variable_1</code> into original dataframe <code>df</code> by concatenating the columns. The last two columns are now the dummy representations of the <code>fuel-type</code> variable. It's all 0s and 1s now.

```python
# Merge dataframes by concatenating the columns
df = pd.concat( [df, dummy_variable_1], axis=1 )
```
