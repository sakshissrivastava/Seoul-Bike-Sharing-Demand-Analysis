# Data Visualization Examples

When working with a new dataset, one of the most useful things to do is to begin to visualize the data. By using tables, histograms, box plots, and other visual tools, we can get a better idea of what the data may be trying to tell us, and we can gain insights into the data that we may have not discovered otherwise.


```python
# import libraries
import seaborn as sns # For plotting and styling
import matplotlib.pyplot as plt # For showing plots and additional customization
import pandas as pd # primary data structure library
import numpy as np # useful for many scientific computing in Python
```

```python
import warnings
# Suppress only specific types of warnings
warnings.filterwarnings('ignore', category=FutureWarning)  # Suppress future deprecation warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)  # Suppress deprecation warnings
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,5) # set the figure width, height (in inches)
sns.set_theme() #use set_theme() if your seaborn is v0.11.0
```

```python
sns.__version__
```

```python
# Load in the data set
tips_data = pd.read_csv("tips.csv")
tips_data.info()
```

```python
# Print out the first few rows of the data
tips_data.head()
```

```python
# Print out the summary statistics for the quantitative variables
tips_data.describe()
```

### Crosstab

```python
# Exploring the time of day x Party Size
pd.crosstab(tips_data['size'], tips_data['time'])
```

```python
# Exploring the time of day x Party Size
pd.crosstab(tips_data['size'], tips_data['time'], normalize='columns')*100
```

```python
# Exploring the time of day x Party Size 
cm = sns.light_palette("green", as_cmap=True)
(round(pd.crosstab(tips_data['size'], tips_data['time'], normalize='columns') * 100,2)).style.background_gradient(cmap = cm)
```

## Histogram

After we have a general 'feel' for the data, it is often good to get a feel for the shape of the distribution of the data.

```python
sns.histplot(tips_data["total_bill"])
```

```python
# Plot a histogram of the total bill
sns.histplot(tips_data["total_bill"]).set_title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot a histogram of the total bill, kde indicates whether to plot a gaussian kernel density estimate.
sns.histplot(tips_data["total_bill"], kde = True).set_title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot a histogram of the Tips only, kde indicates whether to plot a gaussian kernel density estimate.
sns.histplot(tips_data["tip"], kde = False).set_title("Histogram of Tip")
plt.show()
```

```python
# Plot a histogram of both the total bill and the tips in the same graph
sns.histplot(tips_data["total_bill"], kde = False)
sns.histplot(tips_data["tip"], kde = False)
plt.title("Histogram of Both Tip and Total Bill")
plt.show()
```

## Boxplot

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. 

Boxplots and Histograms often complement each other and help an analyst get more information about the data.

```python
# Create a boxplot of the total bill, orient 'v'|'h' indictes Orientation of the plot 
sns.boxplot(data = tips_data["total_bill"]).set_title("Boxplot of Total Bill")
plt.show()
```

```python
# Create a boxplot of the tips
sns.boxplot(data = tips_data["tip"]).set_title("Boxplot of Tip")
plt.show()
```

```python
# Create a boxplot of the tips
sns.boxplot(data = tips_data, x ="tip", y = "time", orient="h").set_title("Boxplot of Tip")
plt.show()
```

## Histograms and Boxplots Plotted by Groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Using graphs, we can see if there is a difference between the tipping amounts of smokers vs. non-smokers, if tipping varies according to the time of the day, or we can explore other trends in the data as well.

```python
# Create a boxplot for total_bill and tip, orient 'v'|'h' indictes Orientation of the plot 
sns.boxplot(data = tips_data[["total_bill", "tip"]], orient="h").set_title("Boxplot of Total Bill")
plt.show()
```

```python
# Create histograms of the tips grouped by time of day
g = sns.FacetGrid(tips_data, col = "time")
g.map(plt.hist, "tip") # Apply a plotting function to each facet's subset of the data
plt.show()
```

```python
# Create histograms of the tips grouped by the day
g = sns.FacetGrid(tips_data, col = "day")
g.map(plt.hist, "tip")
plt.show()
```

```python
# Plot hitograms of Tip, grouped by day and time
g = sns.FacetGrid(tips_data, row='day', col='time')
g.map(plt.hist, 'tip')
g.add_legend()
plt.show()
```

```python
# Create a boxplot of the tips grouped by smoking status
sns.boxplot(data = tips_data, x="smoker", y="tip").set_title("Boxplot of Tips, grouped by smoking status")
plt.show()
```

```python

# Create a boxplot of the tips grouped by time of day
sns.boxplot(data = tips_data, x="time", y="tip").set_title("Boxplot of Tips, grouped by time of day")
plt.show()
```

```python
# Create a boxplot of the tips grouped by day
sns.boxplot(data = tips_data, x="day", y="tip")
plt.title("Boxplot of Tips, grouped by day")
plt.show()
```

```python
# Create a boxplot of the tips grouped by day and smoker
sns.boxplot(data = tips_data, x="day", y="tip", hue="smoker")
plt.title("Boxplot of Tips, grouped by day")
plt.show()
```

## Scatter plot

```python
# create a scatterplot of tip and total bill
sns.scatterplot(data = tips_data, x="tip", y="total_bill").set_title("Scatterplot of Tips and Total Bill")
plt.show()
```

```python
# Use color cue to indicate smoker
sns.scatterplot(data = tips_data, x="tip", y="total_bill", hue='smoker')
plt.title("Scatterplot of Tips and Total Bill")
plt.show()
```

```python
# Use both color cue and shape to indicate smoker
sns.scatterplot(data = tips_data, x="tip", y="total_bill", hue='smoker', style='smoker')
plt.title("Scatterplot of Tips and Total Bill")
plt.show()
```

```python
# Add symbol size to indicate party size. 

# The following two lines of code are equivalent, they are two different ways to write the plot 
#sns.scatterplot(x=tips_data['tip'], y=tips_data['total_bill'], hue=tips_data['smoker'], style=tips_data['smoker'], size=tips_data['size'])
sns.scatterplot(data=tips_data, x='tip', y='total_bill', hue='smoker', style='smoker', size='size')

plt.title("Scatterplot of Tips and Total Bill")
plt.show()
```

```python
# Showing multiple relationships with facets
sns.relplot(data=tips_data, x="total_bill", y="tip", hue="smoker",style='smoker', col="time")
plt.show()
```

```python
# Plotting pairwise data relationships for ALL numeric columns in the dataframe
g = sns.PairGrid(tips_data)
g.map_diag(sns.histplot) # histogram for univariate 
g.map_offdiag(sns.scatterplot) # scatter plot for bivariate 
plt.show()
```
